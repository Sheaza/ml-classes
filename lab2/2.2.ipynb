```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
def function(x):
    return 5*x**2 + 3*x - 6
```

```python
def function_g(x):
    return (np.e**x)/(np.e**x+1)
```

```python
x = np.linspace(-1, 1, 20)
y = function(x)
y2 = function_g(x)
```

```python
sns.lineplot(x=x, y=y)
sns.lineplot(x=x, y=y2)
plt.xlabel('x')
plt.ylabel('y')
```
