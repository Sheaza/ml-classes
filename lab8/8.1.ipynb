```python
import pandas as pd
```

```python
data = pd.read_csv("flats_for_clustering.tsv", sep='\t')
```

```python
data1 = data.dropna()
data1.loc[data1["Piętro"] == "parter", "Piętro"] = 0
data1.loc[data1["Piętro"] == "niski parter", "Piętro"] = 0
data1.loc[data1["Piętro"] == "poddasze", "Piętro"] = -1
```

```python
X = data1
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
from sklearn.cluster import KMeans

model = KMeans(n_clusters=5, verbose=1)
```

```python
model.fit(X)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = X.copy()
X_pca = pca.fit_transform(X_pca)

model = KMeans(n_clusters=5, verbose=1)
model.fit(X_pca)
```

```python
import seaborn as sns

sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=model.labels_, palette=['g', 'r', 'c', 'm', 'y'], s=10)
sns.scatterplot(x=model.cluster_centers_[:, 0], y=model.cluster_centers_[:, 1], hue=range(5), palette=['g', 'r', 'c', 'm', 'y'], s=70, ec="black", marker='X', legend=False)
```
