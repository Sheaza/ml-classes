```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-1, 1, 20)
y = np.linspace(-1, 1, 20)

x, y = np.meshgrid(x, y)
```

```python
def function(x, y):
    return -(x**2 + y**3)
```

```python
z = function(x, y)
```

```python
fig = plt.figure(figsize=(7,7))
ax = plt.axes(projection="3d")
ax.plot_surface(X=x,Y=y,Z=z)
plt.xlabel("x")
plt.ylabel("y")
```
